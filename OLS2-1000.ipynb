```python
# importing required libraries
import os, sys
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math as m
```

```python
def compute(i,rpehe):
    path = "csv_files/1000_"
    train_data = pd.read_csv(path+'train'+str(i)+'.csv')
    test_data = pd.read_csv(path+'test'+str(i)+'.csv')
    train_y = pd.read_csv(path+'train_yf'+str(i)+'.csv')
    train_cf = pd.read_csv(path+'train_ycf'+str(i)+'.csv')
    train_z = pd.read_csv(path+'train_t'+str(i)+'.csv')
    test_y = pd.read_csv(path+'test_yf'+str(i)+'.csv')
    test_z = pd.read_csv(path+'test_t'+str(i)+'.csv')
    train_data['treatment']=train_z
    test_data['treatment']=test_z
    train_data['y']=train_y
    
    control_train_data = train_data[train_data.treatment==0]
    treated_train_data = train_data[train_data.treatment==1]
    
    control_train_X = control_train_data.drop(columns=['treatment','y'],axis=1)
    control_train_y = control_train_data[['y']]
    treated_train_X = treated_train_data.drop(columns=['treatment','y'],axis=1)
    treated_train_y = treated_train_data[['y']]
    
    #control model
    reg_control = LinearRegression().fit(control_train_X,control_train_y)
    #treated model
    reg_treated = LinearRegression().fit(treated_train_X,treated_train_y)
    test_data = test_data.drop(columns=['treatment'])
    #test
    #find control prediction
    control_predicted_test = reg_control.predict(test_data)
    #find treated prediction
    treated_predicted_test = reg_treated.predict(test_data)
    test_data['control_predicted']=control_predicted_test
    test_data['treated_predicted']=treated_predicted_test
    ite=test_data['treated_predicted']-test_data['control_predicted']
    actual_ite = pd.read_csv(path+'test_ite'+str(i)+'.csv')
    pehe = ((ite-actual_ite.iloc[:,0])**2).mean()
    rmse = m.sqrt(pehe)
    print(pehe)
    print(rmse)
    rpehe[i] = rmse
    
    train_data = train_data.drop(columns=['treatment','y'])
    train_iet = np.where(train_z==1,train_y, train_cf)
    #find control prediction
    control_predicted_test = reg_control.predict(train_data)
    #find treated prediction
    treated_predicted_test = reg_treated.predict(train_data)
    ite=treated_predicted_test-control_predicted_test
    pehe = ((ite-train_iet)**2).mean()
    rmse = m.sqrt(pehe)
    print('train', pehe)
    print('train',rmse)
    
   
    return rpehe
```

```python
rpehe = np.zeros(1000)
for i in range(1000):
    rpehe = compute(i,rpehe)
```
