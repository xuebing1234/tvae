```python
# importing required libraries
import os, sys
import pandas as pd
import numpy as np
import math as m
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.stats import sem
```

```python
def compute(i,rpehe):
    path = "csv_files/1000_"
    train_data = pd.read_csv(path+'train'+str(i)+'.csv')
    test_data = pd.read_csv(path+'test'+str(i)+'.csv')
    train_y = pd.read_csv(path+'train_yf'+str(i)+'.csv')
    train_z = pd.read_csv(path+'train_t'+str(i)+'.csv')
    train_data['treatment']=train_z
    reg = LinearRegression().fit(train_data, train_y)
    
    control_x = train_data.copy()
    control_x['treatment']=0
    control_pred = reg.predict(control_x)

    #treatment group
    treatment_x = train_data.copy()
    treatment_x['treatment']=1
    treatment_pred = reg.predict(treatment_x)

    train_data['control_pred']=control_pred
    train_data['treatment_pred']=treatment_pred
    
    ite = train_data['treatment_pred']-train_data['control_pred'].mean()
    test_data['ite']=ite
    actual_ite = pd.read_csv(path+'test_ite'+str(i)+'.csv')
    pehe = ((ite-actual_ite.iloc[:,0])**2).mean()
    rmse = m.sqrt(pehe)
    print(pehe)
    print(rmse)
    rpehe[i] = rmse
    return rpehe
```

```python
rpehe = np.zeros(1000)
for i in range(1000):
    rpehe = compute(i,rpehe)
```
